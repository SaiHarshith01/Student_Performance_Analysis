```python
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['math score'], kde=True)
plt.title("Distribution of Math Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.savefig("images/score_distribution.png", bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='gender', y='math score', data=df)
plt.title("Gender vs Math Scores")
plt.savefig("images/gender_vs_scores.png", bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title("Test Preparation vs Math Scores")
plt.savefig("images/test_prep_vs_scores.png", bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(
    df[['math score', 'reading score', 'writing score']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap of Scores")
plt.savefig("images/correlation_heatmap.png", bbox_inches="tight")
plt.show()
```
